```python
import requests
from bs4 import BeautifulSoup

year = 2021


codes = []
for year in range(2021, 2023+1):
    # '리스트가 없습니다.'라는 colspan=9 속성이 안나올때까지 순회하게 됨. 
    page_data = None
    page = 1
    while page_data is None:
        url = f"http://www.ipostock.co.kr/sub03/ipo02.asp?str4={year}&str5=all&page={page}"
        page += 1
        req = requests.get(url, headers={"User-Agent": get_user_agents()})
        soup = BeautifulSoup(req.content, "lxml", from_encoding="utf-8")

        page_data = soup.select_one('td[colspan="9"]')
        if page_data is not None:
            break
        

        a_tags = soup.select('a[href*="view_pg"][class=""]')

        for tag in a_tags:
            href = tag['href']
            code = href.split('code=')[1].split('&')[0]
            codes.append(code)
len(codes)
```

```python
# 특년년도의 1개 페이지의 기업 목록에 있는 모든 code를 배열로 담음. 
year=2022
page=1

url = f"http://www.ipostock.co.kr/sub03/ipo02.asp?str4={year}&str5=all&page={page}"
req = requests.get(url)
soup = BeautifulSoup(req.content, "lxml", from_encoding="utf-8")


a_tags = soup.select('a[href*="view_pg"][class=""]')

codes = []
for tag in a_tags:
    href = tag['href']
    code = href.split('code=')[1].split('&')[0]
    codes.append(code)
print(codes)
```
