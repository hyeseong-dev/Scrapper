```python
import pymysql

def connect_to_db(host, user, password, db):
    conn = pymysql.connect(
        host=host,
        user=user,
        password=password,
        db=db,
        charset='utf8mb4',
        cursorclass=pymysql.cursors.DictCursor
    )

    return conn
```
